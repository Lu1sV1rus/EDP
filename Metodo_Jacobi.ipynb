<p><font size="4">&nbsp;&nbsp;&nbsp;&nbsp;Codigo para matrices 3x3</p>

<p><font color="lime" size="4">
&nbsp;&nbsp;&nbsp;&nbsp;
Que se debe de cambiar</font><font size="3">
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"M" es la matriz que contiene los valores de X
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"R" s la matriz que contiene los valores de O
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"error" es una variable que detiene el proceso cuando la diferencia en los valores es minima
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"kk" es nuestra variable de conteo</p>

```python
#Datos
M=[(25,-0.9,-0.3),(3.7,-7.3,-0.1),(0.7,0.1,-8.2)]
R=[20.2,-18.9,-56.4]
error=1e-5
x1=x2=x3=0
kk=12

print(0," |          x1          |          x2          |          x3           ")
for k in range(kk):
  a1=a2=a3=0         #Limpiamos las variables
  #Calculamos las xi
  a1=(R[0]-x2*M[0][1]-x3*M[0][2])/M[0][0]
  a2=(R[1]-x1*M[1][0]-x3*M[1][2])/M[1][1]
  a3=(R[2]-x1*M[2][0]-x2*M[2][1])/M[2][2]
  k=k+1   #Actualizamos el contador
  print(k," | ",a1," | ",a2," | ",a3)
  if abs(x1-a1)<error:            #Comprobamos la tolerancia
    print("El numero converge")
    break
  if k==kk:              #Comprobamos que aun no excedemos el limite
    print("Numero max de iteraciones, el procedimiento no fue exitoso")
  x1=a1         #Actualizamos las xi
  x2=a2
  x3=a3
```

<p><font color="lime" size="3">&nbsp;&nbsp;&nbsp;&nbsp;Como podemos ver, nos regresa una lista con los valores de X </font></p>



<p><font size="6">&nbsp;&nbsp;&nbsp;&nbsp;
Por otro lado para matrices nxn </font></p>

<p><font color="lime" size="4">
&nbsp;&nbsp;&nbsp;&nbsp;
Que se debe de cambiar</font><font size="3">
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"M" es la matriz que contiene los valores de X
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"R" es la matriz que contiene los valores de O
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"error" es una variable que detiene el proceso cuando la diferencia en los valores es minima
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"kk" es nuestra variable de conteo
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"x" es la matriz que contendra a las x, su tamano depende del numero de renglones en M
</p><p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
"a" es la matriz que contendra a las x anteriores, su tamano depende del numero de renglones en M</p>

```python
#Datos
M=[(25,-0.9),(3.7,-7.3)]
R=[20.2,-18.9]
error=1e-5
x=[0,0]
kk=12

print(0," |               Los valores de X             |")
for k in range(kk):
  a=[0,0]  #No olvidar modificar
  for i in range(len(R)):
    dem=0   #Limpiamos las variables
    num=0
    for j in range(len(R)):
      if i==j:        #Separamos al numerador aii
        num=M[i][j]
      else:
        dem=dem+(x[j]*M[i][j]) #Realizamos la sumatoria de 1 a n
    a[i]=(R[i]-dem)/num         #Determinamos xi
  print(k+1," | ",a," | ")
  if abs(x[0]-a[0])<error:      #Comprobamos la tolerancia
    print("El numero converge")
    break
  if k==kk:              #Comprobamos que aun no excedemos el limite
    print("Numero max de iteraciones, el procedimiento no fue exitoso")
  x=a               #Actualizamos las x
  k=k+1             #Actualizamos el contador
```

<p><font size="3">Como podemos ver, nos regresa una lista con los valores de X, en orden y separados por comas, el codigo es mas largo,</font><font color="lime" size="3"> pero puede resolver Matrices nxn </font></p>
