Graficar:

$$ y = x^2 $$

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,0.1)
y = x*x
h="Graficar y=x^2"
plt.plot(x,y, color='green')
plt.xlabel('X',color='green')
plt.ylabel('Y',color='green')
plt.title(h, color='green')
plt.show()
```
